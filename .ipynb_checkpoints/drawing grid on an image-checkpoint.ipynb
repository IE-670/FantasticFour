```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image = mpimg.imread("campus2.png")
plt.imshow(image);
```

```python
image.shape
```

```python
x_cor = np.random.randint(image.shape[0], size=(10,))
x_cor
```

```python
y_cor = np.random.randint(image.shape[1], size=(10,))
y_cor
```

```python
len(x_cor)
```

```python
x_cor[1]
```

```python
pts = [[x_cor[0], y_cor[0]]]
pts
```

```python
for i in range(1, len(x_cor)):
    pts.append([x_cor[i], y_cor[i]])

pts
```

```python
pts_array = np.array(pts)
```

```python
plt.imshow(image)
plt.scatter(pts_array[:, 1], pts_array[:, 0], marker="x", color="red", s=200)
plt.show()
```
